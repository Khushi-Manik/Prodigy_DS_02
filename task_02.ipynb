```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train.csv')

print(data.head())
print(data.info())
```

```python
print(data.isnull().sum())

sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title('Missing Data Heatmap')
plt.show()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)

data.drop(columns=['Cabin'], inplace=True)

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

print(data.isnull().sum())
```

```python
print(data.describe())
```

```python
sns.histplot(data['Age'], kde=True, bins=30, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

sns.countplot(x='Sex', data=data, palette='pastel')
plt.title('Gender Distribution')
plt.show()

sns.countplot(x='Survived', data=data, palette='cool')
plt.title('Survival Distribution (0 = Not Survived, 1 = Survived)')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=data, palette='muted')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=data, palette='Blues')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=data, palette='coolwarm')
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
